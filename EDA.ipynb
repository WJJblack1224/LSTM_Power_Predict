```python
import pandas as pd

record_counts = {}

for i in range(1, 18):  # 從 L1 到 L17
    # 讀取資料
    file_path = f'/content/gdrive/My Drive/太陽能發電預測/L{i}_Train.csv'
    df = pd.read_csv(file_path)

    # 計算每個資料集的筆數
    record_counts[f'L{i}'] = len(df)

# 將結果輸出
for dataset, count in record_counts.items():
    print(f'{dataset}: {count} records')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(6, 3, figsize=(20, 30))  # 建立6行3列的子圖布局
fig.suptitle('Distribution of Power (mW) Across 17 Datasets', fontsize=20)

for i in range(1, 18):
    # 讀取資料
    file_path = f'/content/gdrive/My Drive/太陽能發電預測/L{i}_Train.csv'
    df = pd.read_csv(file_path)

    row = (i - 1) // 3
    col = (i - 1) % 3

    sns.histplot(df['Power(mW)'], bins=10, kde=True, ax=axes[row, col])

    axes[row, col].set_title(f'Dataset L{i}')
    axes[row, col].set_xlabel('Power (mW)')
    axes[row, col].set_ylabel('Frequency')

for j in range(17, 18):
    row = j // 3
    col = j % 3
    fig.delaxes(axes[row, col])

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(6, 3, figsize=(25, 35))  
fig.suptitle('Correlation Matrices of 17 Datasets', fontsize=20)  

for i in range(1, 18):  
    file_path = f'/content/gdrive/My Drive/太陽能發電預測/L{i}_Train.csv'
    df = pd.read_csv(file_path)

    #只保留要繪製關聯矩陣的欄位
    df = df[["WindSpeed(m/s)","Pressure(hpa)","Temperature(°C)","Humidity(%)","Sunlight(Lux)","Power(mW)"]]
  
    correlation_matrix = df.corr()

    row = (i - 1) // 3
    col = (i - 1) % 3

    # 繪製關聯矩陣的熱圖
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, ax=axes[row, col])

    # 設定子圖的標題
    axes[row, col].set_title(f'Dataset L{i}')

for j in range(17, 18):  
    row = j // 3
    col = j % 3
    fig.delaxes(axes[row, col])

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
